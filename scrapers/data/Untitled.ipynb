```python
import json
import pandas as pd

# ✅ Charge le fichier — remplace par ton nom de fichier exact
with open("jumia_2026-03-30T10-27-09+00-00.json", encoding="utf-8") as f:
    data = json.load(f)

df = pd.DataFrame(data)

# ==============================
# 1. INFOS GÉNÉRALES
# ==============================
print("=== INFOS GÉNÉRALES ===")
print(f"Total produits    : {len(df)}")
print(f"Colonnes          : {list(df.columns)}")

# ==============================
# 2. PRODUITS PAR CATÉGORIE
# ==============================
print("\n=== PRODUITS PAR CATÉGORIE ===")
print(df["category"].value_counts())

# ==============================
# 3. VALEURS MANQUANTES
# ==============================
print("\n=== VALEURS MANQUANTES ===")
print(df.isnull().sum())

# ==============================
# 4. STATISTIQUES DES PRIX
# ==============================
print("\n=== STATISTIQUES PRIX (en Dhs) ===")
print(df["price"].describe().round(2))

# ==============================
# 5. PRIX PAR CATÉGORIE
# ==============================
print("\n=== PRIX MOYEN PAR CATÉGORIE ===")
print(df.groupby("category")["price"].mean().round(2))

# ==============================
# 6. PRODUITS SANS PRIX
# ==============================
missing_price = df[df["price"].isnull()]
print(f"\n=== PRODUITS SANS PRIX : {len(missing_price)} ===")
if len(missing_price) > 0:
    print(missing_price[["name", "category", "price"]].head(5))

# ==============================
# 7. PRODUITS SANS NOM
# ==============================
missing_name = df[df["name"] == ""]
print(f"\n=== PRODUITS SANS NOM : {len(missing_name)} ===")
if len(missing_name) > 0:
    print(missing_name[["name", "category", "price"]].head(5))

# ==============================
# 8. EXEMPLES PAR CATÉGORIE
# ==============================
print("\n=== 3 EXEMPLES PAR CATÉGORIE ===")
for cat in df["category"].unique():
    print(f"\n--- {cat.upper()} ---")
    print(df[df["category"] == cat][["name", "price", "url"]].head(3).to_string())

# ==============================
# 9. VÉRIFICATION URLS
# ==============================
print("\n=== VÉRIFICATION URLS ===")
urls_valides = df["url"].str.startswith("https://www.jumia.ma").sum()
print(f"URLs valides : {urls_valides}/{len(df)}")

# ==============================
# 10. APERÇU FINAL
# ==============================
print("\n=== APERÇU FINAL ===")
df.head(10)
```

```python
# ✅ Supprimer les produits sans prix
df_clean = df.dropna(subset=["price"])

print(f"Avant nettoyage : {len(df)} produits")
print(f"Après nettoyage : {len(df_clean)} produits")
print(f"Supprimés       : {len(df) - len(df_clean)} produits")

# ✅ Sauvegarder le fichier propre
df_clean.to_json(
    "jumia_clean.json",
    orient="records",
    force_ascii=False,
    indent=2
)

print("\n✅ Fichier sauvegardé : jumia_clean.json")
```

```python
import json
import pandas as pd

fichiers = {
    "jumia_clean.json",
    "micromagma_2026-03-30T21-48-05+00-00.json",
    "zara_2026-03-31T10-08-12+00-00.json",
}

all_data = []
for fichier in fichiers:
    with open(fichier, encoding="utf-8") as f:
        data = json.load(f)
        all_data.extend(data)
        print(f"✅ {fichier} → {len(data)} produits")

df = pd.DataFrame(all_data)

# ✅ Supprimer les doublons
df = df.drop_duplicates(subset=["url"])
df = df.dropna(subset=["price"])

print(f"\n=== RÉSUMÉ FINAL ===")
print(f"Total produits : {len(df)}")
print(f"\nPar catégorie :")
print(df["category"].value_counts())
print(f"\nPar source :")
print(df["source"].value_counts())

# ✅ Sauvegarder
df.to_json(
    "all_products.json",
    orient="records",
    force_ascii=False,
    indent=2
)
print("\n✅ Fichier fusionné sauvegardé : all_products.json")
```
